```python
import tensorflow as tf
from tensorflow.keras.models import load_model
import pickle
import pandas as pd
import numpy as np
```

```python
## Load the ANN train model,Scaler pickle and onehot
model = load_model('model.h5')

## Load the encoder and scaler
with open ('label_encoder_gender.pkl','rb') as file:
    label_encoder_gender = pickle.load(file)

with open ('onehot_encoder_geo.pkl','rb') as file:
    onehot_encoder_geo = pickle.load(file)

with open ('Scaler.pkl','rb') as file:
    Scaler = pickle.load(file)
```

```python
## Example Input Data
input_data = {
    'CreditScore':600,
    'Geography':'France',#This is also a categorical data
    'Gender':'Male', # This is a categorical Data So we will have to convert it to the numerical data with the help of Encoders
    'Age':40,
    'Tenure':3,
    'Balance':60000,
    'NumOfProducts':2,
    'HasCrCard':1,
    'IsActiveMember':1,
    'EstimatedSalary':50000
}
```

```python
##One hot Encoding Geography
geo_encoded = onehot_encoder_geo.transform([[input_data['Geography']]]).toarray()
geo_encoded_df = pd.DataFrame(geo_encoded,columns=onehot_encoder_geo.get_feature_names_out(['Geography']))
geo_encoded_df
```

```python
## The data is now in the key value pair so now we have to convert it into the dataframe
input_df = pd.DataFrame([input_data])
input_df
```

```python
## Encode cetegorical variables  => Gender column
input_df['Gender'] = label_encoder_gender.transform(input_df['Gender'])
input_df
```

```python
## Concatination =>Combine this one-hot encoded columns with input data
input_df = pd.concat([input_df.drop("Geography",axis = 1),geo_encoded_df],axis= 1)
input_df
```

```python
## Scaling the input data
input_scaled = Scaler.transform(input_df)
input_scaled
```

# Model Prediction

```python
prediction = model.predict(input_scaled)
prediction
```

```python
prediction_prob = prediction[0][0]
prediction_prob
```

```python
if prediction_prob > 0.5:
    print("The customer is likely to churn/Exit.")
else:
    print("The customer is not likely to Churn/Exit.")
```
